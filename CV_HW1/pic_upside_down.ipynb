```python
a = np.zeros((4, 4, 3), dtype=np.int)

b = np.random.randint(0,10,size=[4,4,3]) #隨機整數矩陣

print(a)

print(b)
```

```python
for j in range(3):
    for i in range(4):
        for k in range(4):
            a[i][k][j] = b[abs(i-4+1)][k][j]
            print(a)
```

```python
for j in range(3):
    for i in range(512):
        for k in range(512):
            storage_matrix[i][k][j] = img[abs(i-512+1)][k][j]
print(storage_matrix)
```

```python
cv2.imwrite('new_image.bmp', storage_matrix)

new_image = cv2.imread('new_image.bmp')

cv2.imshow('new_image',new_image)

cv2.waitKey(0)
```
